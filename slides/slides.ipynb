## Slide 3

This is a python code plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("penguins")
sns.set_theme(style = "white", palette = "Pastel1")
g = sns.relplot(
data = df, 
x = "bill_length_mm", 
y = "flipper_length_mm", 
hue = "species",
col = "island",
s = 200
)
plt.setp(g._legend.get_texts(), fontsize = 18)
plt.setp(g._legend.get_title(), fontsize = 18)
for i in range(3):
    g.axes[0, i].set_xlabel("Penguin bill length (mm)", fontsize = 18) 
g.axes[0, 0].set_ylabel("Penguin flipper length (mm)", fontsize = 18)
g.axes[0, 0].set_title("Torgersen Island", fontsize = 18)
g.axes[0, 1].set_title("Biscoe Island", fontsize = 18)
g.axes[0, 2].set_title("Dream Island", fontsize = 18)
for i in range(3):
    g.axes[0, i].set_xticks([40, 50, 60])
    g.axes[0, i].set_xticklabels([40, 50, 60], fontsize = 18)
g.axes[0, 0].set_yticklabels([0, 170, 180, 190, 200, 210, 220, 230], fontsize = 18)
plt.savefig("plot-1.pdf", bbox_inches = "tight")
```
